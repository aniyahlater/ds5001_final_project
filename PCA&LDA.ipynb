# PCA Models

## Generating the Document Component Matrix

```python
pca_engine = PCA(n_components=3)
DCM = pd.DataFrame(pca_engine.fit_transform(TFIDF), index=TFIDF.index)
DCM.columns = ['PC{}'.format(i) for i in DCM.columns]
DCM
```

The number of components is 3.

## Component Term Matrix

```python
COMPS = pd.DataFrame(pca_engine.components_.T * np.sqrt(pca_engine.explained_variance_))
COMPS.columns = ["PC{}".format(i) for i in COMPS.columns]
COMPS.index = TFIDF.columns
COMPS.index.name = 'term_str'
COMPS.T
```

### Getting the top 5 positive and negative terms 

```python
top_terms = []
data = []
for i in range(3):
    for j in [0, 1]:
        comp_str = ' '.join(COMPS.sort_values(f'PC{i}', ascending=bool(j)).head(5).index.to_list())
        top_terms.append((f"PC{i}", j, comp_str))
        
COMP_GLOSS = pd.DataFrame(top_terms).set_index([0,1]).unstack()
COMP_GLOSS.index.name = 'comp_id'
COMP_GLOSS.columns = COMP_GLOSS.columns.droplevel(0) 
COMP_GLOSS = COMP_GLOSS.rename(columns={0:'positive', 1:'negative'})
COMP_GLOSS.columns.name = 'polarity'
COMP_GLOSS
```

# LDA Models

```python
# importing the necessary packages 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation as LDA, NMF 
from sklearn.feature_extraction import text
```

```python
# first going to read in corpus data frame 
OHCO = ['mag_id', 'page_num','token_num']
SENTENCE = OHCO[:2],
MAGAZINE = OHCO[:1]
BAG = MAGAZINE
TOKENS = pd.read_csv('/Users/aniyahmcwilliams/Spring 2026/Text As Data/final project/DATA/CORPUS.csv', index_col=['mag_id','page_num','token_num'])
TOKENS.head()
```

```python
# converting the tokens table back to DOC style
DOCS = TOKENS[TOKENS.pos.str.match(r'^NNS?$')]\
    .groupby(BAG).term_str\
    .apply(lambda x: ' '.join(map(str,x)))\
    .to_frame()\
    .rename(columns={'term_str':'doc_str'})
```

# Creating the Vector Space


```python
my_stop_words = list(text.ENGLISH_STOP_WORDS.union(['yes']))
```

```python
count_engine = CountVectorizer(
    stop_words='english',
    max_df=0.5,
    min_df=4,
)
count_model = count_engine.fit_transform(DOCS.doc_str)
TERMS = count_engine.get_feature_names_out()
VOCAB = pd.DataFrame(index=TERMS)
VOCAB.index.name = 'term_str'
DTM = pd.DataFrame(count_model.toarray(), index=DOCS.index, columns=TERMS)
DTM
```

```python
DTM.to_csv('/Users/aniyahmcwilliams/Spring 2026/Text As Data/final project/LDA/DTM.csv')
```

# Generating the Model

```python
n_topics = 5
max_iter = 1000
n_top_terms = 10
TNAMES = [f"T{str(x).zfill(len(str(n_topics)))}" for x in range(n_topics)]
topic_engine = LDA(n_components=n_topics, max_iter=max_iter, random_state=42)
topic_model = topic_engine.fit_transform(count_model)
```

## Calculating THETA

```python
THETA = pd.DataFrame(topic_model, index=DOCS.index, columns=TNAMES)
THETA.columns.name = 'topic_id'
THETA.sample(10).T.style.background_gradient(axis=None)
```

```python
THETA.to_csv('/Users/aniyahmcwilliams/Spring 2026/Text As Data/final project/LDA/THETA.csv')
```

## Calculating PHI

```python
PHI = pd.DataFrame(topic_engine.components_, columns=TERMS, index=TNAMES)
PHI.index.name = 'topic_id'
PHI.columns.name = 'term_str'
PHI.T.sample(10).T.style.background_gradient(axis=None)
```

```python
PHI.to_csv('/Users/aniyahmcwilliams/Spring 2026/Text As Data/final project/LDA/PHI.csv')
```

## Getting the Topics 

```python
pd.set_option('display.max_colwidth', 200)
TOPICS = PHI.stack().groupby('topic_id')\
    .apply(lambda x: ' '.join(x.sort_values(ascending=False).head(n_top_terms).reset_index().term_str))\
    .to_frame('top_terms')
TOPICS
```

```python
#saving every table for LDA
TOPICS.to_csv('/Users/aniyahmcwilliams/Spring 2026/Text As Data/final project/LDA/TOPICS.csv')
```
